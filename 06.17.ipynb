```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/body/body.csv')
df.head()
```

```python
min = df['이완기혈압(최저) : mmHg'].mean()
max = df['수축기혈압(최고) : mmHg'].mean()
result = max - min
print(result)
```

```python
target = df[(df['측정나이'] >= 50) & (df['측정나이'] <= 59)]
result = target['신장 : cm'].mean()
print(result)
```

```python
df['target'] = df['측정나이'] // 10 * 10
result = df['target'].value_counts()
print(result)
```

```python
result = df.groupby(['target','등급'],as_index=False).size()
print(result)
```

```python
ma = df[(df['측정회원성별'] == 'M') & (df['등급'] == 'A')]['체지방율 : %'].mean()
md = df[(df['측정회원성별'] == 'M') & (df['등급'] == 'D')]['체지방율 : %'].mean()
result = md - ma
print(result)
```

```python
ma = df[(df['측정회원성별'] == 'F') & (df['등급'] == 'A')]['체중 : kg'].mean()
md = df[(df['측정회원성별'] == 'F') & (df['등급'] == 'D')]['체중 : kg'].mean()
result = md - ma
print(result)
```

```python
df['bmi'] = df['체중 : kg'] / (df['신장 : cm']/100) ** 2
target = df[df['측정회원성별'] == 'M']['bmi'].mean()
print(target)
```

```python
target = df[(df['체지방율 : %']) > (df['bmi'])]['체중 : kg'].mean()
print(target)
```

```python
md = df[df['측정회원성별'] == 'M']['악력D : kg'].mean()
fd = df[df['측정회원성별'] == 'F']['악력D : kg'].mean()
result = md - fd
print(result)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/scipy2.csv')
df.head()
```

```python
from scipy.stats import levene
from scipy.stats import fligner
from scipy.stats import bartlett

a = df[df['class'] == 'A']
b = df[df['class'] == 'B']

print(levene(a['score'],b['score'],center='median'))
print(levene(a['score'],b['score'],center='mean'))
print(fligner(a['score'],b['score'],center='median'))
print(fligner(a['score'],b['score'],center='mean'))
print(bartlett(a['score'],b['score']))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/ind2.csv')


plt.hist(df1,label='df1',alpha=0.4)
plt.hist(df2,label="df2",alpha=0.4)
plt.xlabel('Score bins')
plt.ylabel('Counts')
plt.legend()
plt.show()
print(df1.shape)
print(df2.shape)
print(df1.head())
```

```python
from scipy.stats import ttest_ind
from scipy.stats import levene
from scipy.stats import fligner
from scipy.stats import bartlett

print(levene(df1['data'], df2['data'],center='median'))
print(levene(df1['data'], df2['data'],center='mean'))
print(fligner(df1['data'], df2['data'],center='median'))
print(fligner(df1['data'], df2['data'],center='mean'))
print(bartlett(df1['data'], df2['data']))
#등분산을 가짐

result = ttest_ind(df1, df2, equal_var=True, alternative='two-sided')
print(result)
#pvalue 값은 0.05보다 작다 귀무가설(평균은 동일하다)기각, 평균은 동일하지 않
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/rel2.csv')

fig ,ax = plt.subplots(1,2)
ax[0].boxplot(df['before'])
ax[1].boxplot(df['after'])
ax[0].set_xticklabels(['before'])
ax[1].set_xticklabels(['after'])
ax[0].set_ylim(100,350)
ax[1].set_ylim(100,350)
ax[1].get_yaxis().set_visible(False)
ax[0].set_ylabel('value')
plt.show()
print(df.head())
```

```python
from scipy.stats import ttest_rel
from scipy.stats import levene
from scipy.stats import fligner
from scipy.stats import bartlett
from scipy.stats import shapiro

print(shapiro(df))
#pvalue 값이 0.05보다 크므로 귀무가설을 기각하지 않는다(정규성을 가진다)

print(levene(df['after'],df['before'],center='median'))
print(levene(df['after'],df['before'],center='mean'))
print(fligner(df['after'],df['before'],center='median'))
print(fligner(df['after'],df['before'],center='mean'))
print(bartlett(df['after'],df['before']))
#pvalue 값이 0.05보다 크므로 귀무가설(등분산을가짐)을 기각할수없음

result = ttest_rel(df['after'],df['before'], alternative='two-sided')
print(result)
#pvalue 값은 0.05보다 작으므로 귀무가설(투약 전후 평균은 같다)를 기각한다. 투약 전후 평균은 다르
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/dice.csv')

plt.bar(df.dice_number,df.counts)
plt.xlabel('dice value')
plt.ylabel('counts')
plt.show()
df.shape
print(df)
```

```python
from scipy.stats import chisquare
df['ratio'] = (df['counts'].sum() / len(df['counts'])).astype('int')
print(df)
result = chisquare(df['counts'],df['ratio'])
print(result)
#pvalue 값이 0.05보다 크므로 귀무가설(각 눈의 확률은 동일하다)를 기각할수 없다 확률은 동일하ㅍ
```

```python
# 주사위 눈금의 발생확률은 1/6으로 모두 동일하다. 그러므로 각 눈금의 기댓값은 실제 발생한 모든값을 6으로 나눈 값이다.

from scipy.stats import chisquare
df['expected'] = (df['counts'].sum()/6).astype('int')
print(chisquare(df.counts,df.expected))

# p-value는 0.8로 귀무가설인 "각 주사위 눈금 발생비율은 동일함"을 기각 할 수 없다
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/hands2.csv')
df.head()
```

```python
from scipy.stats import chisquare
target = df.value_counts().to_frame()
print(target)
target['ratio'] = [int(target['count'].sum() * 0.8), int(target['count'].sum() * 0.2)]
print(target)
result = chisquare(target['count'],target['ratio'])
print(result)
#pvalue값은 0.05보다 작으므로 귀무가설(왼손 오른손 비율이 적합하다)을 기각한다, 왼손 오른손 비율은 적합하지 않
```
