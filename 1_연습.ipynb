```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv', index_col=0)
print(df.head())
print(df.info())
print(df.describe())
```

```python
df_count = df.groupby('channelId').count()
print(df_count)
```

```python
answer =list(df.loc[df.channelId.isin(df.channelId.value_counts().head(10).index)].channelTitle.unique())
print(answer)
```

```python
df_dis = df[df['dislikes'] > df['likes']]
dis_ID = df_dis['channelTitle']
print(list(dis_ID.unique()))
```

```python
df_drop = df[['channelId','channelTitle']].drop_duplicates().channelId.value_counts()
print(df_drop)
target = df_drop[df_drop > 1]
print(len(target))
```

```python
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
answer = df[df['trending_date2'].dt.day_name() == 'Sunday'].categoryId.value_counts()
print(answer.index[0])
```

```python
df['days'] = df['trending_date2'].dt.day_name()
days = df.groupby(['days', 'categoryId'], as_index=False).size()
print(days)
answer = days.pivot(index='categoryId', columns='days')
print(answer)
```

```python
df_comment = df[df.view_count != 0]

df_comment['values'] = df_comment['comment_count'] / df_comment['view_count']
result = df_comment.sort_values(by='values', ascending = False)
print(result['title'].iloc[0])
```

```python
result = df_comment.sort_values(by='values', ascending = True)
result2 = result[result['values'] > 0]
print(result2['title'].iloc[0])
```

```python
df_l_d = df[(df['likes'] != 0) & (df['dislikes'] != 0)]

df_l_d['ratio'] = df_l_d['dislikes'] / df_l_d['likes']
result = df_l_d.sort_values(by='ratio', ascending = True)
print(result['title'].iloc[0])
```

```python
df_many = df.groupby('channelTitle').size().sort_values(ascending=False)
print(df_many.index[0])
```

```python
df_tw = df[['title', 'channelTitle']].sum().sort_values(by='title', ascending=False)
print(df_tw)
```

```python
import pandas as pd

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
df1 = df.groupby('Country').sum().sort_values('Goals', ascending=False)
df1 = pd.DataFrame(df1)
df1.head(5)
```

```python
df2 = df.groupby('Country').size().sort_values(ascending=False)
df2.head(5)
```

```python
df['yearLst'] = df.Years.str.split('-')

def checkFour(x):
  for value in x:
    if len(str(value)) != 4:
      return False
  return True

df['check'] = df['yearLst'].apply(checkFour)

result = len(df[df['check'] == False])
print(result)
```

```python
df2 = df[df['check'] == True].reset_index(drop=True)
print(df2.shape[0])
```

```python
print(df2)
```

```python
df2['LenCup'] = df2['yearLst'].str.len()
print(df2['LenCup'])
result = df2['LenCup'].value_counts()
print(result[4])
```

```python
result = len(df2[(df2['Country'] == 'Yugoslavia') & (df2['LenCup'] == 2)])
print(result)
```

```python
result = len(df2[df2['Years'].str.contains('2002')])
print(result)
```

```python
result = len(df2[df2['Player'].str.lower().str.contains('carlos')])
print(result)
```

```python
result = (df2[df2['LenCup'] == 1]).sort_values('Goals', ascending=False)
print(result['Player'].iloc[0])
```
