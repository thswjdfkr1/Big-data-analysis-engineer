```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df1 = df.copy()
df1.head()
```

```python
answer =list(df.loc[df.channelId.isin(df.channelId.value_counts().head(10).index)].channelTitle.unique())
print(answer)
```

```python
ans = list(df1.loc[df1.dislikes > df1.likes].channelTitle.unique())
print(ans)
```

```python
change = df1[['channelTitle', 'channelId']].drop_duplicates().channelId.value_counts()
ans = change[change>1]
print(len(ans))
```

```python
df1['trending_date2'] = pd.to_datetime(df1['trending_date2'])
ans = df1.loc[df1['trending_date2'].dt.day_name() == 'Sunday'].categoryId.value_counts().index[0]
print(ans)
```

```python
group = df1.groupby([df1['trending_date2'].dt.day_name(),'categoryId'],as_index=False).size()
ans = group.pivot(index='categoryId',columns='trending_date2')
print(ans)
```
