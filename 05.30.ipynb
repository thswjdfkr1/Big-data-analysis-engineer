```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')

df1 = df.copy()
df1.head()
```

```python
ans = df1.groupby('Country').sum().sort_values('Goals', ascending=False).head(5)
print(ans['Goals'])
```

```python
ans = df1.groupby('Country').size().sort_values(ascending=False).head()
print(ans)
```

```python
df1['YearsL'] = df1['Years'].str.split('-')

def check(x):
  for year in x:
    if len(str(year)) != 4:
      return 'F'

    return 'T'

df1['YearsC'] = df1['YearsL'].apply(check)

print(len(df1[df1['YearsC'] == 'F']))
```

```python
df2 = df1[df1['YearsC'] == 'T'].reset_index(drop=True)
print(df2.shape[0])
```

```python
df2['LenCup'] = df2['YearsL'].str.len()
ans = df2['LenCup'].value_counts()[4]
print(ans)
```

```python
ans = df2[df2['Country'] == 'Yugoslavia'].value_counts('LenCup')[2]
print(ans)
```

```python
ans = df2[df2['Years'].str.contains('2002')]
print(len(ans))
```

```python
ans = df2[df2['Player'].str.lower().str.contains('carlos')]
print(len(ans))
```

```python
ans = df2[df2['LenCup'] == 1].sort_values('Goals',ascending=False).Player.values[0]
print(ans)
```

```python
ans = df2[df2['LenCup'] == 1].Country.value_counts().sort_values(ascending=False).index[0]
print(ans)
```
