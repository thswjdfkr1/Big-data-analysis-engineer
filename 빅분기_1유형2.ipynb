```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv",encoding='latin',sep=';')
print(df.head())
```

```python
df_Tor = df[df['Tm'] == 'TOR']
print(df_Tor)

result = round(df_Tor['Age'].mean(), 4)
print(result)
```

```python
df_Age_min = df[df['Age'] == df['Age'].min()]
result = df_Age_min.Pos.value_counts()
print(result.index[0])
```

```python
result = df[df['Age']==df['Age'].min()].Pos.value_counts()
print(result)
```

```python
df_name = df['Player'].str.split().str[0].str.lower().value_counts().head(1)
result = df_name
print(result)
```

```python
df_Pts = df['PTS'].groupby(df['Pos']).mean().sort_values()
print(df_Pts)
```

```python
df_G = df.groupby('Tm')['G'].max().mean()
print(df_G)
```

```python
result = df.sort_values(['Tm','G']).groupby('Tm').tail(1).G.mean()
print(result)
```

```python
df_0 = df[(df['Tm'] == 'MIA') & (df['Pos'].isin(['C','PF']))]
result = df_0['MP'].mean()
print(result)
```

```python
mean_g = df['G'].mean()

result = df[df['G'] >= mean_g * 1.5]['3P'].mean()
print(result)
```

```python
mean_age = df.Age.mean()

df_upper = df[df['Age'] >= mean_age]
df_lower = df[df['Age'] < mean_age]

result = df_upper['G'].mean() - df_lower['G'].mean()
print(result)
```

```python
df_age = df.groupby('Tm')['Age'].mean().sort_values().index[0]
print(df_age)
```

```python
df_pos = df.groupby('Pos')['MP'].mean()
print(df_pos)
```
