```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/normal3.csv')

print(df.head())
print(df.shape)
print(df.describe())
print(df.info())
```

```python
log_df = np.log1p(df)
result = shapiro(log_df)
print(result)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/height1.csv')

print(df.shape)
print(df.head())
print(df.info())
print(df.describe())
```
