```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/데이터 셋/P230601.csv', encoding='utf-8')

print(df.info())
print(df.describe())
print(df.head(1))
```

```python
df['출동일시'] = pd.to_datetime(df['출동일시'])
df['신고일시'] = pd.to_datetime(df['신고일시'])
df['소요시간'] = (df['출동일시'] - df['신고일시']).dt.total_seconds()
print(df['소요시간'])

result = df.groupby([df['출동소방서'],
                     df['신고일시'].dt.year,
                     df['신고일시'].dt.month]).mean('소요시간')
result = result.sort_values(by='소요시간', ascending = False)

print(round(float(result['소요시간'][0] / 60)))
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score


df1 = pd.read_csv('/content/drive/MyDrive/데이터 셋/P210204-01.csv')
print(df1.info())
print(df1.describe())

df1['Reached.on.Time_Y.N'] = df1['Reached.on.Time_Y.N'].astype('category')

df2 = pd.read_csv('/content/drive/MyDrive/데이터 셋/P210204-02.csv')
print(df1.info())
print(df1.describe())
```

```python
x = df1.drop('Reached.on.Time_Y.N', axis=1)
y = df1['Reached.on.Time_Y.N']

x_encode = pd.get_dummies(x)

x_train, x_vaild, y_train, y_vaild = train_test_split(x_encode.drop('ID', axis=1), \
                                                      y, test_size=0.25)

md = RandomForestClassifier(n_estimators=300)
md.fit(x_train, y_train)

pred = md.predict(x_vaild)
```

```python
cm = confusion_matrix(y_vaild, pred, labels=[1,0])
print(cm)
print(accuracy_score(y_vaild, pred))
print(precision_score(y_vaild, pred))
print(f1_score(y_vaild, pred))
print(roc_auc_score(y_vaild, pred))
```

```python
x_test = df2
x_test_encoded = pd.get_dummies(x_test)

md = RandomForestClassifier(n_estimators=300)
md.fit(x_encode.drop('ID',axis=1), y)

pred = md.predict_proba(x_test_encoded.drop('ID', axis=1))
print(pred)

result = pd.DataFrame({'ID':df2['ID'], 'pred':pred[:, 1]})
print(result)
result.to_csv("tngjaqjsgh",index=False)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

df1 = pd.read_csv('/content/drive/MyDrive/데이터 셋/P210304-01.csv')
print(df1.info())
print(df1.describe())
print(df1.head())

df2 = pd.read_csv('/content/drive/MyDrive/데이터 셋/P210304-02.csv')
print(df2.info())
print(df2.describe())
print(df2.head())
```

```python
x = df1.drop('TravelInsurance', axis=1)
y = df1['TravelInsurance']

x_encoded = pd.get_dummies(x)

x_train, x_valid, y_train, y_valid = train_test_split(x_encoded.drop('X', axis =1), \
                                                     y, test_size=0.25)

md = RandomForestClassifier(n_estimators=300)
md.fit(x_train, y_train)

pred = md.predict(x_valid)
```

```python
cm = confusion_matrix(y_valid, pred, labels=[1,0])
print(cm)
print(accuracy_score(y_valid, pred))
print(precision_score(y_valid, pred))
print(recall_score(y_valid, pred))
print(f1_score(y_valid, pred))
print(roc_auc_score(y_valid, pred))
```

```python
x_test = df2
x_test_encoded = pd.get_dummies(x_test)

md = RandomForestClassifier(n_estimators=300)
md.fit(x_encoded.drop('X', axis=1), y)

pred = md.predict_proba(x_test_encoded.drop('X', axis=1))
print(pred)

result = pd.DataFrame({'index': df2['X'], 'y_pred':pred[:,1]})
print(result)

result.to_csv("2", index=False)
```
