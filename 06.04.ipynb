```python
import pandas as pd
df1 =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bicycle/seoul_bi.csv')
df = df1.copy()

print(df.head())
print(df.shape)
```

```python
target = df['대여일자'].value_counts().sort_index(ascending=True).to_frame()
result = target[target['count'] == target['count'].max()].index[0]
print(result)
```

```python
df['day_name'] = pd.to_datetime(df['대여일자']).dt.day_name()
ans = df['day_name'].value_counts().sort_values(ascending=False).to_frame()
print(ans)
```

```python
ans = df.groupby(['day_name','대여소번호']).size().to_frame('size').sort_values(['day_name','size'],ascending=False).reset_index()
result = ans.drop_duplicates('day_name',keep='first').reset_index(drop=True)
print(result)
```

```python
daily = df[df['대여구분코드'].isin(['일일권','일일권(비회원)'])].연령대코드.value_counts().sort_index()
total = df['연령대코드'].value_counts().sort_index()

ratio = (daily/total).sort_values(ascending=False).index[0]
print('가장 높은 비율의 나이대:',ratio)
```

```python
target = df[['연령대코드','이동거리']].groupby('연령대코드').mean()
print(target)
```

```python
ans = df[df['연령대코드'] == '20대'].reset_index(drop=True)
mean = ans[ans['이동거리'] > ans['이동거리'].mean()].reset_index(drop=True)
target = mean.sort_values(['대여일자','대여소번호'],ascending=False).reset_index(drop=True).iloc[0:200].탄소량

target = target.astype('float')
result = round(target.sum()/len(target),3)
print(result)
```

```python
ans = df[(df['대여일자'] == '2021-06-07') & (df['연령대코드'] == '~10대')].이용건수.median()
print(ans)
```

```python
df['대여일자'] = pd.to_datetime(df['대여일자'])
df['day_name'] = df['대여일자'].dt.day_name()

ans = df[(df['day_name'].isin(['Tuesday', 'Wednesday', 'Thursday', 'Friday','Monday'])) &(df['대여시간'].isin([6,7,8]))]

result = ans.groupby(['대여시간','대여소번호']).size().to_frame('이용 횟수')

answer = result.sort_values(['대여시간','이용 횟수'],ascending=False).groupby('대여시간').head(3)
display(answer)
print(display(answer))
```

```python
ans = df[df['이동거리'] > df['이동거리'].mean()].reset_index(drop=True)

result = ans['이동거리'].std()
print(result)
```

```python
df['sex'] = df['성별'].map(lambda x: '남' if x in ['M','m'] else '여')

result = df[['sex','이동거리']].groupby('sex').mean()
print(result)
```
