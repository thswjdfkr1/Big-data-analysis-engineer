```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/x_test.csv")

x_train = x_train.drop(['id'], axis=1)
x_test = x_test.drop(['id'], axis=1)

display(x_train.head())
display(y_train.head())
```

```python
x_df = pd.merge(x_train, y_train, on='ID')
```

```python
print(x_df.head(), '\n')
print(x_df.shape, '\n')
print(x_df.info(), '\n')
print(x_df.describe(), '\n')
print(x_df.isnull().sum())
```

```python
print(x_test.head(), '\n')
print(x_test.shape, '\n')
print(x_test.info(), '\n')
print(x_test.describe(), '\n')
print(x_test.isnull().sum())
```

```python
x = x_df.drop(['ID','Response'], axis=1)
y = x_df['Response']
```

```python
obj_col = x.select_dtypes('object').columns
num_col = x.select_dtypes('number').columns

for col in obj_col:
  print(col)
  print(x[col].unique(), '\n')
```

```python
dum_col = obj_col

x = pd.get_dummies(x, columns=dum_col, dtype=int)
x_test = pd.get_dummies(x_test, columns=dum_col,dtype=int)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
x[num_col] = mms.fit_transform(x[num_col])
x_test[num_col] = mms.transform(x_test[num_col])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x,y,test_size=0.2, stratify=y,random_state=810)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

params = {'n_estimators':[5,10,15,20],'max_depth':[5,10,15,20]}
gscv = GridSearchCV(estimator=rf, param_grid=params, cv=5)
gscv.fit(x_train,y_train)

print(gscv.best_params_)
```

```python
model = RandomForestClassifier(max_depth=10, n_estimators=15, random_state=810)
model.fit(x_train,y_train)

pred_t = model.predict(x_val)
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score,confusion_matrix

print(accuracy_score(y_val, pred_t))
print(confusion_matrix(y_val,pred_t))
print(recall_score(y_val,pred_t))
print(f1_score(y_val,pred_t))
```

```python
target_x = x_test.drop(['ID'], axis=1)
pred = model.predict(target_x)

result = pd.DataFrame({'ID':x_test['ID'],'pred':pred})
result.to_csv(',수험번호.csv',index=False)
print(result)
```

```python
df1 = df.drop('나라명',axis=1)
target = df1[df1['년도'] == 2019].corr().unstack().to_frame().reset_index().dropna()
result = target[target[0] != 1].sort_values(0, ascending = False).head(5)
result.columns = (['v1','v2','corr'])
print(result)
```

```python
target = df.groupby('년도').tail(5)
result = target.groupby('년도')['점수'].mean()
print(result)
```

```python
target = df[df['년도']==2019]
mean = target['상대GDP'].mean()
mean_up = target[target['상대GDP'] >= mean]['점수'].mean()
mean_down = target[target['상대GDP'] <= mean]['점수'].mean()
result = mean_up - mean_down
print(result)
```
