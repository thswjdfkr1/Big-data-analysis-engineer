```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/consum/Tetuan%20City%20power%20consumption.csv')
df.head()
```

```python
df['Datetime'] = pd.to_datetime(df['DateTime'])
result = df['Datetime'].dt.month.value_counts().sort_index().to_frame()
print(result)
```

```python
df['Datetime']= pd.to_datetime(df['Datetime'])
target = df[df['Datetime'].dt.month == 3]
result = target.groupby(target['Datetime'].dt.hour)['Temperature'].mean().sort_values(ascending=True).iloc[0]
print(result)
```

```python
target = df[df['Zone 1 Power Consumption'] > df['Zone 2  Power Consumption']]
result = target['Humidity'].mean()
print(result)
```
