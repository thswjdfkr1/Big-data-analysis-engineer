```python
x = x_df.drop(['ID','charges'], axis=1)
y = x_df[['charges']]
print(x)
print(y)
```

```python
num_col = x.select_dtypes(exclude='object').columns
obj_col = x.select_dtypes(include='object').columns

for col in obj_col:
  print(col)
  print(x[col].unique(), '\n')
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x[num_col] = mms.fit_transform(x[num_col])
x_test[num_col] = mms.transform(x_test[num_col])
```

```python
x = pd.get_dummies(x, columns = obj_col, dtype=int)
x_test = pd.get_dummies(x_test, columns = obj_col, dtype=int)
```

```python
print(x)
print(x_test)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.3, random_state=810)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=810)

params = {'n_estimators':[30,35,40,45,50], 'max_depth':[5,10,15,20]}

gscv = GridSearchCV(estimator= rf, param_grid=params,cv=5)
gscv.fit(x_train, y_train)
print(gscv.best_params_)
```

```python
model = RandomForestRegressor(max_depth=5, n_estimators=45, random_state=810)
model.fit(x_train, y_train)
pred_t = model.predict(x_val)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print('mse:', mean_squared_error(y_val, pred_t))
print('mae:', mean_absolute_error(y_val, pred_t))
print('rmse:', (mean_squared_error(y_val, pred_t)) ** 0.5)
print('r2_score:', r2_score(y_val, pred_t))
```

```python
target = x_test.drop(['ID'], axis=1)
pred = model.predict(target)

result = pd.DataFrame({'ID':x_test['ID'], 'pred':pred})
result.to_csv('수험번호.csv',index=False)
print(result)
```
