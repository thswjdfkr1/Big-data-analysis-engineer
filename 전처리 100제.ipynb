```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv')
```

```python
df.shape
df.head()
```

```python
df1 = df.copy()
df1.head()
```

```python
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +' : 'e'
}

df1['newincome'] = df1.Income_Category.map(lambda x: dic[x])
ans = df1['newincome']

print(ans)
```

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':
        return 'b'
    elif x =='$60K - $80K':
        return 'c'
    elif x =='$80K - $120K':
        return 'd'
    elif x =='$120K +' :
        return 'e'

df['newIncome']  =df.Income_Category.apply(changeCategory)

Ans = df['newIncome']
```

```python
df1['AgeState'] = df1['Customer_Age'].map(lambda x: x//10 * 10)
ans = df1['AgeState'].value_counts().sort_values(ascending=True).sort_index()
print(ans)
```

```python
def Level(x):
  if x in 'Graduate':
    return 1
  else:
    return 0

df1['newEduLevel'] = df1['Education_Level'].apply(Level)
ans1 = df1['newEduLevel'].value_counts()
print(ans1)
```

```python
df1['newEduLevel'] = df1.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
Ans = df1['newEduLevel'].value_counts()

print(Ans)
```

```python
df1['newlimit'] = df1['Credit_Limit'].map(lambda x: 1 if x >= 4500 else 0)
ans = df1['newlimit'].value_counts()
print(ans)
```

```python
def check(x):
  if x.Marital_Status == 'Married' and x.Card_Category == 'Platinum':
    return 1
  else:
    return 0

df1['newstate'] = df1.apply(check, axis=1)
ans = df1['newstate'].value_counts()
print(ans)
```

```python
def check(x):
  if x == 'M':
    return 'male'
  else:
      return 'female'

df1['Gender'] = df1['Gender'].apply(check)
ans = df1['Gender'].value_counts()
print(ans)
```
