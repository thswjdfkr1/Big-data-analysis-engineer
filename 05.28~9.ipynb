```python
import pandas as pd

channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')

print(channel.head())
print('---------')
print(video.head())
```

```python
video['ct']= pd.to_datetime(video['ct'])

ans = video['videoname'].value_counts()
print(ans)
```

```python
video['ct'] = video['ct'].sort_values(ascending=False)

print(video[['viewcnt','videoname','ct']])
```

```python
channel['ct'] = pd.to_datetime(channel['ct'])

target = channel[channel['ct'] >= pd.to_datetime('2021-10-3')].sort_values('ct').drop_duplicates('channelname')
ans = target[['channelname','subcnt']].reset_index(drop=True)
print(ans)
```

```python
end = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-11-01 15']
start = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-10-03 03']

end_df = end[['channelname','subcnt']].reset_index(drop=True)
start_df = start[['channelname','subcnt']].reset_index(drop=True)

end_df.columns = ['channelname','end_cnt']
start_df.columns = ['channelname','start_cnt']

df1 = pd.merge(start_df, end_df, on='channelname')
df1['del'] = df1['end_cnt'] - df1['start_cnt']

ans = df1[['channelname','del']]
print(ans)
```

```python
start_date = video.sort_values(['ct','videoname']).drop_duplicates('videoname',keep='first')[['ct','videoname']]
start_date['date'] = start_date['ct'].dt.date

ans = start_date[['date','videoname']]
print(ans)
```

```python
video['time'] = video.ct.dt.hour

ans = video.loc[video['time'] == 21].sort_values(['videoname','ct']).drop_duplicates('videoname'). \
sort_values('viewcnt', ascending=False)[['videoname','viewcnt','ct']].reset_index(drop=True)
print(ans)
```

```python
target = video.sort_values('ct',ascending=False).drop_duplicates('videoname',keep='first')

target['ratio'] = target['dislikecnt'] / target['likecnt']
ans = target.sort_values('ratio',ascending=True)[['videoname','ratio']].reset_index(drop=True)
print(ans)
```

```python
start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')

target = video.loc[(video['ct'] >= start) & (video['ct'] <= end)].reset_index(drop=True)

def check(x):
  result = max(x) - min(x)
  return result

ans = target[['videoname','viewcnt']].groupby('videoname').agg(check)

print(ans)
```
