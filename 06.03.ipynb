```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/dice.csv')

df1 = df.copy()
df1.head()
```

```python
from scipy.stats import chisquare

df1['ratio'] = (df1['counts'].sum()) / 6

result = chisquare(df1['counts'],df1['ratio'])
print(result)

#pvalue가 0.8, 귀무가설을 기각할수 없고 각 눈금의 발생 비율은 동일함
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/hands2.csv')
df_a = df.copy()
print(df_a.head(3))
```

```python
from scipy.stats import chisquare

target = df_a['hands'].value_counts().to_frame(name='count')
total_count = target['count'].sum()

target['ratio'] = [int(total_count*0.8),int(total_count*0.2)]

result = chisquare(target['count'],target['ratio'])
print(target)
print(result)

#pvalue 값이 0.05보다 작으므로 귀무가설을 기각, 왼손 오른손 02 08은   적합하지 않음
```
