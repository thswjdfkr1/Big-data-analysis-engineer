```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.shape,'\n')
print(train.head(3),'\n')
print(train.info(),'\n')
print(train.describe(),'\n')
print(train.isnull().sum())
```

```python
print(test.shape,'\n')
print(test.head(3),'\n')
print(test.info(),'\n')
print(test.describe(),'\n')
print(test.isnull().sum())
```

```python
x = train.drop(['ID','연월','이용금액'],axis=1)
y = train['이용금액']
target = test.drop(['ID','연월'],axis=1)
```

```python
obj_col = x.select_dtypes('object').columns
for col in obj_col:
  print(col)
  print(x[col].unique(),'\n')
```

```python
obj_col = target.select_dtypes('object').columns
for col in obj_col:
  print(col)
  print(x[col].unique(),'\n')
```

```python
le_col = obj_col.drop('성별')
from sklearn.preprocessing import LabelEncoder
for col in le_col:
  le = LabelEncoder()
  x[col] = le.fit_transform(x[col])
  target[col] = le.transform(target[col])
```

```python
x = pd.get_dummies(x, columns=['성별'], dtype=int)
```

```python
target = pd.get_dummies(target, columns=['성별'], dtype=int)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms_col = x.select_dtypes(exclude='object').columns
x[mms_col] = mms.fit_transform(x[mms_col])
target[mms_col] = mms.transform(target[mms_col])
```

```python
print(x.head(1))
print(target.head(1))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=810)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=810)
params = {'n_estimators':range(2,12), 'max_depth':range(2,7)}
gscv = GridSearchCV(estimator=rf, param_grid=params, cv=5)
gscv.fit(x_train, y_train)
print(gscv.best_params_)
```

```python
model = RandomForestRegressor(max_depth=6, n_estimators=11, random_state=810)
model.fit(x_train, y_train)
pred1 = model.predict(x_val)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('mse:',mean_squared_error(y_val, pred1))
print('rmse:',mean_squared_error(y_val, pred1)**0.5)
print('rmse:',mean_squared_error(y_val,pred1, squared=False))
print('r2_score:',r2_score(y_val,pred1))
```

```python
pred = model.predict(target)
result = pd.DataFrame({'ID':test['ID'],'pred':pred})
result.to_csv('수험번호.csv', index=False)
print(result)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_2_.csv')
df.head(5)
```

```python
df['BMI'] = df['weight(kg)'] / ((df['height(cm)'] / 100) ** 2)

def type(x):
  if x > 25:
    return('초고도')
  elif 25 > x >= 23:
    return('고도')
  elif 23 > x >= 18.5:
    return('정상')
  elif 18.5 > x:
    return('저체중')
df['type'] = df['BMI'].apply(type)

result = df[(df['type']=='초고도') + (df['type']=='저체중')]
print(len(result))
```

```python
import numpy as np
n = 100
p_hat = 60/100
p=0.5
a=0.05
z = (p_hat - p) / np.sqrt(p*(1-p)/n)
print(round(z,5))
```

```python
from scipy.stats import norm
result = 1-norm.cdf(z)
print(round(result,3))
#pvalue가 0.05보다 작으므로 귀무가설을 기각한다, 평균점수는 50점 이상이다
```

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.shape,'\n')
print(train.head(3),'\n')
print(train.info(),'\n')
print(train.describe(),'\n')
print(train.isnull().sum())
print(train['General_Health'].unique())
```

```python
print(test.shape,'\n')
print(test.head(3),'\n')
print(test.info(),'\n')
print(test.describe(),'\n')
print(test.isnull().sum())
```

```python
x = train.drop(['ID','General_Health'],axis=1)
y = train['General_Health']
target = test.drop('ID',axis=1)
```

```python
obj_col = x.select_dtypes(include='object').columns
for col in obj_col:
  print(col)
  print(x[col].nunique(),'\n')
```

```python
obj_col = target.select_dtypes(include='object').columns
for col in obj_col:
  print(col)
  print(x[col].nunique(),'\n')
```

```python
mms_col = x.select_dtypes(exclude='object').columns

from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x[mms_col] = mms.fit_transform(x[mms_col])
target[mms_col] = mms.transform(target[mms_col])
```

```python
le_col = ['Checkup','Age_Category']

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in le_col:
  x[col] = le.fit_transform(x[col])
  target[col] = le.transform(target[col])
```

```python
dum_col = obj_col.drop(le_col)

x = pd.get_dummies(x, columns=dum_col, dtype=int)
target = pd.get_dummies(target, columns= dum_col, dtype=int)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x,y, test_size=0.2, stratify=y, random_state=810)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=810)
params = {'n_estimators':range(2,12), 'max_depth':range(5,15)}
gscv = GridSearchCV(estimator=rf, param_grid=params, cv=5)
gscv.fit(x_train, y_train)
print(gscv.best_params_)
```

```python
model = RandomForestClassifier(max_depth=6, n_estimators=11, random_state=810)
model.fit(x_train,y_train)
pred1 = model.predict(x_val)
```

```python
from sklearn.metrics import f1_score

print('f1_score:',f1_score(y_val,pred1, average='macro'))
```

```python
pred = model.predict(target)
result = pd.DataFrame({'ID':test['ID'], 'pred':pred})
result.to_csv('수험번호.csv', index=False)
print(result)
```
