```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/pok/Pokemon.csv')
df.head()
print(df.shape)
```

```python
target = df.groupby('Legendary')['HP'].mean()
result = target.values[1] - target.values[0]
print(result)
```

```python
target = df['Type 2'].value_counts().sort_values(ascending=False)
result = target.index[0]
print(result)
```

```python
target = df['Type 1'].value_counts().sort_values(ascending=False)
ans = target.index[0]
print(ans)
target1 = df[df['Type 1'] == 'Water']
print(target1)
result = (target1['Attack'].mean()) / (target1['Defense'].mean())
print(result)
```

```python
target = df[df['Legendary'] == True].value_counts('Generation').sort_values(ascending=False)
print(target)
result = target.index[0]
print(result)
```

```python
target = df[['HP', 'Attack', 'Defense','Sp. Atk', 'Sp. Def', 'Speed']].corr().unstack().reset_index().rename(columns={0:'corr'})
result = target[target['corr'] != 1].sort_values('corr', ascending=False).reset_index(drop=True)
print(result.iloc[0])
```

```python
target = df.sort_values(['Generation','Attack']).groupby('Generation').head(3).Attack.mean()
print(target)
```

```python
target = df.sort_values(['Generation','Attack'],ascending=False).groupby('Generation').head(5).Attack.mean()
print(target)
```

```python
target = df[['Type 1','Type 2']].value_counts()
print(target)
result = target.head(1)
print(result)
```

```python
target = df[['Type 1','Type 2']].value_counts()
result = target[target ==1]
print(len(result))
```

```python
target = df[['Type 1','Type 2']].value_counts()
target1 = target[target==1]

lst = []
for value in target1.reset_index().values:
  t1 = value[0]
  t2 = value[2]

  sp = df[(df['Type 1']==t1) & (df['Type 2']==t2)]
  lst.append(sp)

result = pd.concat(lst).reset_index(drop=True).Generation.value_counts()
print(result)
```

```python
target = df[['Type 1','Type 2']].value_counts()
target2 =target[target==1]

lst = []
for value in target2.reset_index().values:
    t1 = value[0]
    t2 = value[1]

    sp = df[(df['Type 1']==t1) & (df['Type 2']==t2)]
    lst.append(sp)

result = pd.concat(lst).reset_index(drop=True).Generation.value_counts().sort_index()
print(result)
```
