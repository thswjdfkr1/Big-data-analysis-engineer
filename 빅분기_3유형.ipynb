```python
import pandas as pd
import numpy as np
from scipy.stats import chisquare

df = pd.read_csv('/content/drive/MyDrive/데이터 셋/P230605.csv', encoding='cp949')

print(df.head())
print(df.shape)
print(df.info())
print(df.describe())
print(df['코드'].value_counts())
```

```python
df_placebo = df.groupby('코드').size().reset_index(name='건수')
print(df_placebo)

total_count = len(df)

df_placebo['유형비율'] = df_placebo['건수'] / total_count
print(df_placebo)

result_1 = round((df_placebo['유형비율'][3]),3)
print(result_1)
```

```python
from scipy.stats import chisquare
import numpy as np

df_placebo['비율']= [0.05, 0.1, 0.05, 0.8]
print(df_placebo)

result2_chiq = chisquare(df_placebo['건수'], f_exp = df_placebo['비율'] * np.sum(total_count))
result2 = round(result2_chiq.statistic, 3)
print(result2)

result3 = round(result2_chiq.pvalue, 3)
print(result3)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df= pd.read_csv('/content/drive/MyDrive/데이터 셋/P230606.csv', encoding='cp949')

print(df.shape)
print(df.head())
print(df.info())
print(df.describe())
```

```python
x= df.drop(['Temperature'], axis=1)
y = df['Temperature']
print(x)
print(y)

md = LinearRegression()
md.fit(x,y)

result1 = round(float(md.coef_[0]),3)
print(result1)
```

```python
from scipy.stats import shapiro,levene, ttest_ind

nor = shapiro(df)
print(nor)

lev = levene(df['Wind'], y)

ttest = ttest_ind(df['Wind'],y)
result2 = round(ttest.pvalue, 3)
print(result2)
```

```python
new = pd.DataFrame({'O3':[10],'Solar': [90], 'Wind':[20]})

result3 = md.predict(new)
print(np.round(result3[0], 3))
```
