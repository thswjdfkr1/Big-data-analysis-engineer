```python
import pandas as pd

x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")
y_test =pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_test.csv")

print(x_train.shape)
print(x_train.head())
print(x_train.info())
print(x_train.describe())
```

```python
print(y_train.shape)
print(y_train.head())
```

```python
print(x_test.shape)
print(x_test.head())

print(y_test.shape)
print(y_test.head())
```

```python
x_train.sort_values(by='CustomerId', ascending=False)
x_test.sort_values(by='CustomerId', ascending=False)
y_train.sort_values(by='CustomerId', ascending=False)
y_test.sort_values(by='CustomerId', ascending=False)

x_train = x_train[['CustomerId','CreditScore', 'Geography', 'Gender', 'Age',
       'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary']]
x_test = x_test [['CustomerId','CreditScore', 'Geography', 'Gender', 'Age',
       'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary']]
y_train = y_train['Exited']
y_test = y_test['Exited']
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse= False)
ohe_x_train = ohe.fit_transform(x_train[['Geography', 'Gender']])
ohe_x_train_df = pd.DataFrame(ohe_x_train, columns=ohe.get_feature_names_out())
print(ohe_x_train_df)

ohe = OneHotEncoder(sparse= False)
ohe_x_test = ohe.fit_transform(x_test[['Geography', 'Gender']])
ohe_x_test_df = pd.DataFrame(ohe_x_test, columns=ohe_x_train_df.columns)
print(ohe_x_test_df)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms_list = ['CreditScore','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']

mms = MinMaxScaler()

x_train[mms_list] = mms.fit_transform(x_train[mms_list])
x_test[mms_list] = mms.fit_transform(x_test[mms_list])

print(x_train)
```

```python
x_train_s = pd.concat([ohe_x_train_df, x_train], axis=1)
print(x_train_s)

x_test_s = pd.concat([ohe_x_test_df, x_test], axis=1)
print(x_test_s)
```

```python
x_train_drop = x_train_s.drop(['Age','Gender','Geography','CustomerId'],axis=1)
x_test_drop = x_test_s.drop(['Age','Gender','Geography','CustomerId'],axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=810)

params = {'n_estimators':[20,25,30,40,50,60,80,100], 'max_depth':[5,10,15,20]}

gs = GridSearchCV(estimator=rf, param_grid=params, cv=3)
gs.fit(x_train_drop, y_train)
print(gs.best_params_)
print(gs.best_score_)
```

```python
rf = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=810)
rf.fit(x_train_drop, y_train)
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score

pred = rf.predict(x_test_drop)
pred_prob = rf.predict_proba(x_test_drop)

print("accuracy_score:", accuracy_score(y_test, pred))
print("roc_auc_score:", roc_auc_score(y_test, pred_prob[:,1]))
```

```python
result = pd.DataFrame({"CustomerId": x_test['CustomerId'].sort_values(), \
                       'Exited': pred})
print(result)
```

```python
result.to_csv('수험번호',index=False)
```

```python
import pandas as pd

train_x = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_train.csv")
train_y = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/y_train.csv")
test_x = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_test.csv")
test_y = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/y_test.csv')

print(train_x.shape)
print(train_x.head())
print(train_x.info())
print(train_x.describe())
```

```python
print(train_x.isnull().sum())
print(test_x.isnull().sum())
```

```python
x_train_nunique = []
# x_test_nunique = []

for col in train_x.select_dtypes(include=object):
     print(col)
     print(train_x[col].unique(), '\n')
     x_train_nunique.append(train_x[col].nunique())

print(x_train_nunique)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe_list = ['Gender','Product_importance','Customer_care_calls','Mode_of_Shipment',\
           'Warehouse_block']

ohe = OneHotEncoder(sparse=False)
ohe_train_x = ohe.fit_transform(train_x[ohe_list])
ohe_train_x_df = pd.DataFrame(ohe_train_x, columns=ohe.get_feature_names_out())

ohe_test_x = ohe.fit_transform(test_x[ohe_list])
ohe_test_x_df = pd.DataFrame(ohe_test_x, columns=ohe_train_x_df.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms_list = ['Customer_rating', 'Cost_of_the_Product', 'Prior_purchases',
       'Discount_offered', 'Weight_in_gms']

mms = MinMaxScaler()

train_x[mms_list] = mms.fit_transform(train_x[mms_list])
test_x[mms_list] = mms.fit_transform(test_x[mms_list])

train_x = train_x.drop(ohe_list, axis=1)
train_x = pd.concat([train_x, ohe_train_x_df], axis=1)

test_x = test_x.drop(ohe_list, axis=1)
test_x = pd.concat([test_x, ohe_test_x_df], axis=1)
```

```python
print(train_x.shape)
print(test_x.shape)
```

```python
train_x = train_x.sort_values(by='ID', ascending=True)
test_x = test_x.sort_values(by='ID', ascending=True)
train_y = train_y.sort_values(by='ID', ascending=True)
test_y = test_y.sort_values(by='ID', ascending=True)

train_x_s = train_x.drop('ID', axis=1)
train_y_s = train_y['Reached.on.Time_Y.N']
print(train_x_s)
print(train_y_s)

test_x_s = test_x.drop('ID', axis=1)
test_y_s = test_y['Reached.on.Time_Y.N']
```

```python
from sklearn.model_selection import train_test_split

tr_x, val_x, tr_y, val_y = train_test_split(train_x_s, train_y_s, stratify=train_y_s, test_size= 0.2)

print(tr_x)
print(val_x)
print(tr_y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, accuracy_score

rf = RandomForestClassifier(random_state=810)
rf.fit(tr_x, tr_y)

test_pred = rf.predict(val_x)
test_pred_prob = rf.predict_proba(val_x)

print("accuracy_score:", accuracy_score(val_y, test_pred))
print("roc_auc_score:", roc_auc_score(val_y, test_pred_prob[:,1]))
```

```python
from sklearn.model_selection import GridSearchCV

params = {'n_estimators':[200,250,300], 'max_depth':[10,20,30,40,50,100]}

gs = GridSearchCV(estimator = rf, param_grid = params, cv=3)
gs.fit(tr_x, tr_y)
print(gs.best_params_)
print(gs.best_score_)
```

```python
rf = RandomForestClassifier(max_depth=10, n_estimators=250, random_state=810)
rf.fit(tr_x, tr_y)
pred = rf.predict(test_x_s)
print(accuracy_score(test_y_s, pred))

result = pd.DataFrame({'ID': test_y['ID'], 'Reached.on.Time_Y.N': pred})
result.to_csv('수험번호.csv', index=False)
```
